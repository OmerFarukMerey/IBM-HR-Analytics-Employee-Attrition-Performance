```python
import pickle
import warnings
import numpy as np
```

```python
np.random.seed(12345)
```

```python
warnings.filterwarnings("ignore")
```

```python
best_model = pickle.load(open("model.pickle",'rb'))
preprocessed_data = pickle.load(open("preprocessing.pickle",'rb'))
```

```python
best_model
```

```python
preprocessed_data
```

```python
from sklearn.model_selection import train_test_split

X = preprocessed_data.drop("Attrition", axis=1)
y = preprocessed_data['Attrition']

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,stratify=y)
```

```python
print(X.shape, y.shape, X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
best_model.fit(X_train, y_train)
```

```python
best_model.score(X_test,y_test)
```

```python
from sklearn.metrics import classification_report, accuracy_score, recall_score, precision_score, f1_score, roc_auc_score
y_preds = best_model.predict(X_test)

print(classification_report(y_test, y_preds))
print()
print("Accuracy Score:", accuracy_score(y_test, y_preds))
print("Recall Score:", recall_score(y_test, y_preds))
print("Precision Score:", precision_score(y_test, y_preds))
print("F1 Score:", f1_score(y_test, y_preds))
print("ROC AUC Score:", roc_auc_score(y_test, y_preds))
```
